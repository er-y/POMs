```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=pd.read_excel('lacunaryPOM.xlsx') #takes experimental data from Yoonrae's Excel file
xval=x['cm**-1']
yval=x['Intensity']
plt.vlines([988.197183,981.222817,962.641591,948.07604],-0.1,1, color='red', label='DFT') #frequencies from orca calculation
plt.hlines([0],599.76244,3997.77341, color='gray', linestyle='dashed') #horizontal line at 0
plt.plot(xval,yval/max(yval), label='Exp')

plt.legend(loc='upper right')
plt.xlabel("Frequency (cm^-1)", size=15)
plt.ylabel("Normalized Intensity", size=15)
plt.title("IR Spectrum of Lacunary POM", size=17)
plt.xlim(3997.77341,599.76244) #limits of experimental data
plt.ylim(-0.1,1.3) #adjusted for readibility of legend and grpah organization
plt.yticks([]) #removes y axis values

plt.tight_layout()
plt.savefig('./IR_LacunaryPOM.pdf')
plt.show()
```
